```
##THIS WORKED FINALLY TO COPY TO COLAB FROM GOOGLE DRIVE

from PIL import Image
import glob
import os
#i=number_files
# new folder path (may need to alter for Windows OS)
# change path to your path
path ='/content/Session2Dataset_final' #the path where to save resized images
# create new folder
if not os.path.exists(path):
    os.makedirs(path)

# loop over existing images and resize
# change path to your path
n=0
#for i in range(cnt):
for dir in subfolders:
  os.chdir(dir) 
  name=os.path.basename(dir)
  fpath=path+'/'+name
  if not os.path.exists(fpath):
    os.makedirs(fpath)
  for filename in glob.glob('*.jpg'): #path of raw images
    n=n+1
    img = Image.open(filename)  #resize((224,224))
    if img.mode != "RGB":
        img = img.convert("RGB")
    # save resized images to new folder with existing filename
    #img.save('{}{}{}'.format(path,'/',os.path.split(filename)[1]))
    #filename=names[i]+str(n)+'.jpg'
    img.save(os.path.join(fpath,filename), 'JPEG', quality=90)

print(n)    
```
